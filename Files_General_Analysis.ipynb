# Identifying and Removing Duplicate rows.

#### Function for identifying that don't have duplicate rows along with the columns that are unique to each file.

```python
def checking_duplicat_entry_files(files):
    files_with_unique_columns = {}
    files_with_duplicate_rows = []
    for key, value in files.items():
        duplicate = count = 0
        shape = value.shape[0]
        columns = value.columns
        for column in columns:
            if value[column].nunique() == shape:
                files_with_unique_columns.setdefault(key,[]).append(column)
                count += 1
            else:
                if count>0:
                    duplicate = 0
                else:
                    duplicate = 1
        if duplicate>0:
            files_with_duplicate_rows.append(key)
    print("Files with Unique columns",files_with_unique_columns)
    print("Files with duplicate rows",files_with_duplicate_rows)
```

```python
files_for_duplicate_rows = {"claims":claims, "encounters":encounters, "patients":patients, "denials":denials}
checking_duplicat_entry_files(files_for_duplicate_rows)
```
