```python
import pandas as pd
import numpy as np
```

# Loading cleaned datasets

```python
claims_and_billing_cleaned = pd.read_excel(r"C:\Users\hp\OneDrive\Documents\Documents\Datasets\Medical claims and denial Data\CA Hospital Dataset – Q1 2025\Data eda\Cleaned Data\claims_and_billing_cleaned.xlsx")
denials_cleaned = pd.read_excel(r"C:\Users\hp\OneDrive\Documents\Documents\Datasets\Medical claims and denial Data\CA Hospital Dataset – Q1 2025\Data eda\Cleaned Data\denials_cleaned.xlsx")
encounters_cleaned = pd.read_excel(r"C:\Users\hp\OneDrive\Documents\Documents\Datasets\Medical claims and denial Data\CA Hospital Dataset – Q1 2025\Data eda\Cleaned Data\encounters_cleaned.xlsx")
patients_cleaned = pd.read_excel(r"C:\Users\hp\OneDrive\Documents\Documents\Datasets\Medical claims and denial Data\CA Hospital Dataset – Q1 2025\Data eda\Cleaned Data\patients_cleaned.xlsx")
```

# Merging datasets

```python
# Identifying common columns in each files.
files_list =  {"claims_and_billing_cleaned":claims_and_billing_cleaned, "denials_cleaned":denials_cleaned, "encounters_cleaned":encounters_cleaned, "patients_cleaned":patients_cleaned}
for key,value in files_list.items():
    print(f"Columns for {key} are",value.columns)
```

```python
print(claims_and_billing_cleaned.shape)
```

```python
# With this we identified that claims_and_billing_cleaned can be merged with encounter on "Encounter_Id" and this can be merged with patients_cleaned on "Patient_Id" and this can be merged with denials_cleaned on "claim_id"

merged_data = claims_and_billing_cleaned.merge(encounters_cleaned[['Encounter_Id', 'Provider_Id', 'Visit_Date', 'Visit_Type', 'Department', 'Reason_For_Visit', 'Diagnosis_Code', 'Admission_Type', 'Discharge_Date', 'Length_Of_Stay', 'Status', 'Readmitted_Flag']], on = "Encounter_Id", how = "left")
merged_data.shape[0]
```

```python
merged_data.columns
```

```python
merged_data = merged_data.merge(patients_cleaned[['Patient_Id', 'Patient_Name', 'Dob', 'Gender', 'Ethnicity',
       'Insurance_Type', 'Marital_Status', 'City', 'State','Registration_Date']], on = "Patient_Id", how = "left")
merged_data.shape[0]
```

```python
merged_data.columns
```

```python
merged_data = merged_data.merge(denials_cleaned[['Claim_Id', 'Denial_Id', 'Denial_Reason_Code',
       'Denial_Reason_Description', 'Denied_Amount', 'Denial_Date',
       'Appeal_Filed', 'Appeal_Status', 'Appeal_Resolution_Date',
       'Final_Outcome']], on = "Claim_Id", how = "left")
merged_data.shape[0]
```

```python
merged_data.isnull().all(axis=1).sum()
```
