# Creating KPI metrics

```python
# Since the final_outcome had three categories, I defined a binary flag where only written-off claims were treated as 1, representing actual revenue loss. Paid and reprocessed claims are considered recoverable, therefore decided to mark them a 0. This is to allow for clearer financial impact. Inorder to do so i am restructing my main_data.
```

```python
main_data.drop(columns=["Final_Status_Flag"], inplace=True)
main_data["Final_Status_Flag"] = main_data["Final_Outcome"].apply(lambda x: 1 if x== "Written off" else 0 )
```

```python
# Reading the file
main_data = pd.read_excel(r"C:\Users\hp\OneDrive\Documents\Documents\Datasets\Medical claims and denial Data\CA Hospital Dataset – Q1 2025\Data eda\Cleaned Data\Final_Merged_data.xlsx")
main_data.head(10)
```

```python
kpi = {}
kpi["Total_Claims"] = main_data["Claim_Id"].nunique()
kpi["Total_Billed_Amount"] = main_data ["Billed_Amount"].sum()
kpi["Total_Paid_Amount"] = round(main_data["Paid_Amount"].sum(),2)
kpi["Written_Off"] = main_data[main_data["Final_Outcome"] == "Written off"]["Denied_Amount"].sum()
kpi["Initial_Denial"] = round(main_data["Claims_Status_Flag"].mean()*100,2)
kpi["Final_Revenue_Loss"] = round(main_data["Final_Status_Flag"].mean()*100,2)
kpi["Appeal_Approval_Rate"] = round(((kpi["Initial_Denial"] - kpi["Final_Revenue_Loss"])/kpi["Initial_Denial"])*100,2)
```

```python
for key, values in kpi.items():
    print(key, values)
```

```python
# Saving the KPI dictionary into xlsx format.
KPI_info = pd.DataFrame(list(kpi.items()),columns=["KPI", "values"])
```

```python
file_path = r"C:\Users\hp\OneDrive\Documents\Documents\Datasets\Medical claims and denial Data\CA Hospital Dataset – Q1 2025\Data eda\Cleaned Data\Final_Merged_data.xlsx"
with pd.ExcelWriter(file_path, mode="a") as writer: KPI_info.to_excel(writer, sheet_name="KPI", index=False)
```
